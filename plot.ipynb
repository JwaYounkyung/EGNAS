```python
def result_plot(val, test, title):
    plt.clf() 
    plt.figure(figsize=(15, 5))

    epochs = range(len(val))

    plt.plot(epochs, val, label='val', marker='o')
    plt.plot(epochs, test, label='test', marker='o')
    plt.xlabel('Generations')
    plt.ylabel('Accuracy')
    
    #plt.title(title)
    plt.legend()

    plt.show()

result_plot(val, test, 'Accuracy')
```
